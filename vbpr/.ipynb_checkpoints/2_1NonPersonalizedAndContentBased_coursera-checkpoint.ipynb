# WEEK ONE
## Introduction
### Recommendation Approaches
- 非个性化、整体流行的
- 物品的相关性
- 基于物品的内容
- 协同推荐（Collabrative)

### Deisigning a Recommender
- 收集观点和实验数据
- 有目的的找相关数据
- 计算推荐
- 以有效的方式展示数据

### Recommenders as Big Data 
- 基于分析和估算的强调：
1. 用数据验证最好的推荐方法
2. 算法不断优化
3. 提高用户体验和满足商业目的
- 持续采用新的机器学习方法

### Personalization Level⭐
- 归类的/热门的：每个人都收到一样的推荐
- 基于人口统计学的：匹配一个目标群体
- 基于短暂的：匹配符合当前活动的
- 基于持续的：匹配长期的兴趣
- 注意不要推荐人们一定会买的东西，不要浪费推荐名额

### Analytical Framework⭐
Dimensions of Analysis:
- 属于领域（大类）(如product)
- 目的(如add on items to buy)
- 推荐的上下文
- 观点的来源(如reviews)
- 个性化的层次（Personalization Level: current products)
- 隐私和可信度
- 外观界面
- 推荐算法

# WEEK TWO
## Introduction
### Why Non-Personlized
- 冷启动
- 简单并且有效
- 围绕公共展示的在线社区
- 可以应用在一些无法个性化的媒体平台

### Breaking it Down
- 欢迎度是一个很重要的度量指标
- 平均值可能误导性强：
    1. 容易被累加喜欢百分比的方式影响（刷分的水军）
    2. 标准化用户的评分（对评分范围的设置要求更高）；
    3. 还要考虑个体评分者的可信度（历史评分，因为有的用户给分一向偏低/偏高）
- 数据越多越好

### Some take-away lessons
- Non-personalized popularity statistics or averages can be effective in the right application:
    Need to understand relationship between average and user need;correct average
- In many cases it can be best to show count, average, and distribution together
- For ranking, one alternative to average is the percentage who score above a threshold (or below)
- Personlization would address many limitations.

### Learning Objectives
- 了解几种计算和展示预测的方法
- 理解如何利用分散的，时间序性的数据来给物品排序
- 理解在预测和推荐的集中

### Ranking Considerations
- Confidence自信度：有多自信推荐的产品是真的好
- Risk tolerance容错率：高风险，高回报；保守的推荐
- Domain and business considerations领域和商业考虑：年龄；系统目标

### How Not To Sort By Average Rating⭐
http://www.evanmiller.org/how-not-to-sort-by-average-rating.html
- WRONG \#1: `Score = (Positive ratings) − (Negative ratings)`<br>
(eg.忽略了占比，只看差)
- WRONG \#2: `Score = Average rating = (Positive ratings) / (Total ratings)`<br>
(eg.忽略了样本总数，只看平均水平，有的商品只有一条评价恰好是五星，有的商品很多条但平均分4.+）
- CORRECT \#3: `Score = Lower bound of Wilson score confidence interval for a Bernoulli parameter`<br>

### Demographics And Related Approaches基于人口统计学和相关的方法
- step1.根据人口统计学的总结统计：eg.在女性/男性人群中最受欢迎的物品
- step2.思考一个多元化的回归模型来预测：eg.线性回归/逻辑回归
- 优势&制衡点：
    - 很多商品和内容的创作基于人口统计学，如电视节目，杂志文章广告，私人用品..
    - 很容易在相对自然的情况下划分目标人群
    - 很容易极度不适合那些不符合对应条件的人群
    
### Product Association Recommenders基于具体产品相关的推荐
- 实时的，基于当前context的个性化
- 如何计算：
    - 版本1：人工制订的捆绑交叉销售表单
    - 版本2：数据挖掘的自动化方法
        - 例如计算$P = \frac{{n}_{买x,y物品}}{{n}_{买x物品}}$但是注意这个算法是不具有对称性的（比如买了x很大可能买y,但买了y不一定很大可能买x)
        - 贝叶斯法则：$P(A|B)=\frac{P(B|A)P(A)}{P(B)}$
    - Instructions-Assignments (for Non-Personalized)
        1. 平均值Mean Rating：计算所有电影的评分平均值，排序并选出最高的三个
        2. 评分数量rating Count(Popularity)：计算每部电影的点评数目，排序并选出top3
        3. 评论4星以上的百分比：计算，排序选出top3
        4. 看过Toy Story并看过最多的电影：用（x+y）/x这个方法，注意是根据评过分的数量计算，选出top3
        5. 跟Toy Story的关联：计算玩具总动员和其他电影的评级向量之间的相关性。可以使用内联函数CORREL()。注意区分共现和共同关联，它们分别反映了不同类型的衡量。
        6. 根据性别的平均评分差异：选出男性和女性中分别评分平均分最高的三部电影
        7. 分别计算男性和女性中四星以上的百分比
