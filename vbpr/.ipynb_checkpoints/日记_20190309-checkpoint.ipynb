最近总结的教训
----
1. **学会拒绝。**要充分理性考虑自己的立场和实际情况，不该答应的要求和没有意义的浪费时间的举动，学会拒绝。学会拒绝。就比如一场聊天，就要学会适可而止，自己心里要有一个度，在话题发展和耗费时间上，不能任由对方牵着走。要自己学会中断。你自己的时间要学会衡量价值，是有成本的，所以需要用到值得的地方。
2. **认识到了自己的完美主义情结。**认识自己没有错，及时改正最关键。自己的完美主义情结主要体现在，开始一项任务，心里就会产生强迫性的情结，要求一定要尽善尽美，从一开始就要系统结构的顾到每一方面。**一旦哪一方面没有做到想象中的完美，心里会产生直接放弃的念头，或者任由其拖延的心理。**这是不正确的。一旦开始有点点错了，后面的心态就开始不好了，想着放弃。比如转专业过来，刚开始的一学期功课没有做好，心里想的不是后面要尽可能补上去，而是产生了破罐子破摔的心态，后面的任由其发展，不想着弥补形势，这实际上是一种缺乏变通思维的表现。心里的观念是必须每一步都要走的正确完美，才值得继续好好的走下去，一旦出了点意料之外的错误，就完全撒手不管，逃避它，企图他自己不了了之。这是不对的。**因为逃避/不解决本身也是一种解决方式，你觉得你暂时没有解决它，但实际上你用逃避的方式解决了它，这样带来的后果就是非常糟糕。不如把脑子变得灵活一点，允许出错，允许出意外，把自己做事情的鲁棒性锻炼出来。**认识到自己的完美主义，并接受它，不必自责，但是学会和自己的完美主义相处，更好的利用它。比如分清轻重缓急，在小事情上没必要完美主义，允许自己去做的不好，但是在一件重要的大事情上，可以允许自己完美主义，去做好计划，做好它。完美主义造成的拖延症，怎么克服呢？就是学会正视自己的缺陷和不完美，直面它，要勇敢地想着怎么去解决它，而不是逃避它。
3. **认识到自己的平庸和局限性。**不要老把自己的期望值设定的太高，自己就是一个普通人，平凡人，有普通人的缺点和欲望，有普通人的惰性和迟钝，不要给自己设定太高的目标，每一天的任务不要期待自己能完成太多。认清自己的实际能力，不要觉得自己就必须惊艳老师，急着让老师认可自己。就把自己想成一个普通学生，在完成普通学生该做的任务就可以了。不要老想着如何让老师更高的评价自己。没用的。做好分内的事情，把该做的事情做好。就算做错了，就算超出时间了，没有关系，你是个普通人，是个平凡人，在老师的眼中，你也是个普通的平庸的学生而已，会犯错，会有拖延毛病，就是正常的印象，不存在自己会很丢脸的情况。厚着脸皮去找老师吧。
4. **学会厚脸皮。**当然是在必要的事情上，学会厚脸皮。比如找工作，比如找导师，比如向别人学习请教东西，比如想跟别人合作，比如在一场亲密关系中想跟对方有更深的发展。学会厚脸皮，学会适当的表达出自己的需求。喜欢一个人，就要让对方意识到你喜欢对方，想跟对方合作，就要让对方意识到你有这个需求和诚意。
