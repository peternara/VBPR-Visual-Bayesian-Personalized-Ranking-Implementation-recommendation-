今天好了不少。
昨天在微信上加了回来，还怀着小小的希望，希望可以和好。但昨晚不出所料，依旧是碰到了冰冷的墙壁上。我说的话，如是像把刀子，已经把别人伤害的太深，也使别人痛苦了一段。把别人吓跑了。不愿意听任何解释的吓跑了。这次我彻底死心了。不管是买的礼物也好，重新联系道歉也好。该尝试的都尝试了。现在别人作何反应，原不原谅都是别人的选择了。由不得我。既然不原谅。那我也死心了。再也没有什么缺少我做的了。所以，我现在倒是释然了很多。也学习了很多。学到了成熟，学到了管住自己的嘴和心情。其他的就任它而去了。没有过不了的坎。该经历的务必经历了。就是这样。
今天看TIS，里面的剧情，也生动给我展示了双方之间一方犯错了，该如何正确的表现。被冒犯方要表现的大度，耐心，给对方一次机会。而做错事的一方，也要及时道歉，及时带有无比诚意的道歉。及时止损。顺着对方的意愿行事，不要再把重心放在自己的身上，关注自己到底委不委屈，还是太幼稚了。既然做错了，就要竭力挽回损失。
TIS第四集，这里面的被收养的女孩，给我示范了，幼稚的表现。
幼稚的人，才会想法通过伤害自己或者伤害别人，来对一件不合心意的事情作出很差劲的反应。这样做的后果是，损人不利己。
